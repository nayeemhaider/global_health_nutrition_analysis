```python
# ==========================================
#   Global Health & Nutrition - EDA Notebook
# ==========================================

from pathlib import Path
import sqlite3
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

sns.set(style="whitegrid")

# ------------------------------------------------
# 1. Auto-detect project root (bulletproof version)
# ------------------------------------------------
def find_project_root(marker="README.md"):
    cur = Path().resolve()
    while cur != cur.parent:
        if (cur / marker).exists():
            return cur
        cur = cur.parent
    raise RuntimeError("Project root not found. Ensure README.md exists in root.")

ROOT = find_project_root()
DB_PATH = ROOT / "data" / "db" / "global_health_nutrition.db"

print("PROJECT ROOT:", ROOT)
print("DATABASE PATH:", DB_PATH)
print("DB Exists:", DB_PATH.exists())
```

```python
# ------------------------------------------------
# 2. Load cleaned view v_core_clean
# ------------------------------------------------
conn = sqlite3.connect(DB_PATH)
df = pd.read_sql_query("SELECT * FROM v_core_clean;", conn)
conn.close()

print("Loaded rows:", len(df))
display(df.head())
```

```python
# ------------------------------------------------
# 3. Basic dataset overview
# ------------------------------------------------
print("Shape:", df.shape)
print("\nColumn preview:\n", df.columns.tolist())
print("\nYears:", df["year"].min(), "→", df["year"].max())
print("Countries:", df["country"].nunique())
print("Genders:", df["gender"].unique())
```

```python
# ------------------------------------------------
# 4. Missing values
# ------------------------------------------------
missing = df.isna().mean().sort_values(ascending=False).to_frame("missing_ratio")
display(missing.head(20))

# Select key variables dynamically (avoid KeyErrors)
requested_cols = [
    "life_expectancy",
    "infant_mortality_rate",
    "under5_mortality_rate",
    "neonatal_mortality_rate",   # may or may not exist
    "uhc_coverage",
    "doctors",
    "nurses_midwifes",
    "basic_water",
    "basic_sanitation_total",
    "safely_sanitation_total",
    "clean_fuel",
    "gdp_per_capita",
]

existing_cols = [c for c in requested_cols if c in missing.index]
print("\nUsing existing cols:", existing_cols)

missing_main = missing.loc[existing_cols].reset_index()
missing_main.columns = ["variable", "missing_ratio"]

plt.figure(figsize=(10, 5))
sns.barplot(data=missing_main, x="variable", y="missing_ratio")
plt.xticks(rotation=45, ha="right")
plt.ylabel("Missing ratio")
plt.title("Missing values in key variables")
plt.tight_layout()
plt.show()
```

```python
# ------------------------------------------------
# 5. Summary statistics by gender
# ------------------------------------------------
display(df.groupby("gender")["life_expectancy"].describe())
display(df.groupby("gender")["infant_mortality_rate"].describe())
```

```python
# ------------------------------------------------
# 6. Global time trends
# ------------------------------------------------
global_trend = (
    df.groupby("year")
      .agg(
          life_expectancy=("life_expectancy", "mean"),
          infant_mortality=("infant_mortality_rate", "mean"),
          under5_mortality=("under5_mortality_rate", "mean"),
      )
      .reset_index()
)

plt.figure(figsize=(10, 5))
plt.plot(global_trend["year"], global_trend["life_expectancy"], label="Life Expectancy")
plt.plot(global_trend["year"], global_trend["under5_mortality"], label="Under-5 Mortality")
plt.legend()
plt.title("Global Trends in Health Outcomes")
plt.xlabel("Year")
plt.tight_layout()
plt.show()
```

```python
# ------------------------------------------------
# 7. Distribution plots
# ------------------------------------------------
fig, axes = plt.subplots(2, 2, figsize=(12, 8))
axes = axes.ravel()

sns.histplot(df["life_expectancy"], bins=30, kde=True, ax=axes[0])
axes[0].set_title("Life expectancy")

sns.histplot(df["infant_mortality_rate"], bins=30, kde=True, ax=axes[1])
axes[1].set_title("Infant mortality rate")

sns.histplot(df["under5_mortality_rate"], bins=30, kde=True, ax=axes[2])
axes[2].set_title("Under-5 mortality rate")

sns.histplot(df["gdp_per_capita"], bins=30, kde=True, ax=axes[3])
axes[3].set_title("GDP per capita")

plt.tight_layout()
plt.show()
```

```python
# ------------------------------------------------
# 8. Correlation heatmap
# ------------------------------------------------
corr_cols = [
    "life_expectancy",
    "infant_mortality_rate",
    "under5_mortality_rate",
    "uhc_coverage",
    "doctors",
    "nurses_midwifes",
    "basic_water",
    "basic_sanitation_total",
    "safely_sanitation_total",
    "clean_fuel",
    "gdp_per_capita",
]

corr_cols = [c for c in corr_cols if c in df.columns]

corr = df[corr_cols].corr()

plt.figure(figsize=(10, 8))
sns.heatmap(corr, annot=False, cmap="coolwarm", center=0)
plt.title("Correlation Matrix of Key Health Indicators")
plt.tight_layout()
plt.show()
```

```python
# ------------------------------------------------
# 9. Scatterplots for relationships
# ------------------------------------------------
latest_year = df["year"].max()
df_latest = df[df["year"] == latest_year]

# Doctors vs life expectancy
if "doctors" in df.columns:
    plt.figure(figsize=(7, 5))
    sns.scatterplot(data=df_latest, x="doctors", y="life_expectancy")
    plt.title(f"Doctors vs Life Expectancy — {latest_year}")
    plt.tight_layout()
    plt.show()

# Sanitation vs under-5 mortality
if "basic_sanitation_total" in df.columns:
    plt.figure(figsize=(7, 5))
    sns.scatterplot(data=df_latest, x="basic_sanitation_total", y="under5_mortality_rate")
    plt.title(f"Sanitation vs Under-5 Mortality — {latest_year}")
    plt.tight_layout()
    plt.show()

# GDP vs life expectancy
plt.figure(figsize=(7, 5))
sns.scatterplot(data=df_latest, x="gdp_per_capita", y="life_expectancy")
plt.xscale("log")
plt.title(f"GDP vs Life Expectancy — {latest_year}")
plt.tight_layout()
plt.show()
```

```python
# ------------------------------------------------
# 10. Outliers
# ------------------------------------------------
display(df_latest.sort_values("life_expectancy").head(10))
display(df_latest.sort_values("life_expectancy", ascending=False).head(10))
```

```python
# ------------------------------------------------
# 11. Export cleaned subset for downstream analysis
# ------------------------------------------------
out_path = ROOT / "data" / "processed" / "global_health_nutrition_clean.csv"
df.to_csv(out_path, index=False)
print("Saved cleaned file to:", out_path)
```
